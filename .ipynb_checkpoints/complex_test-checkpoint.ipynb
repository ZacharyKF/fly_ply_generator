```python
n = 5;
def f_at(j, n):
    f(x) = (cos((2 * j * pi)/n) + I * sin((2 * j * pi)/n))**x;
    return f;

r(x) = sum((f_at(j, n) for j in range(n + 1)))
plot((r(x)).real(), (x, 0, n)).show()
```

```python
n = 10;
j = 1;
f(x) = (cos((2 * j * pi)/n) + I * sin((2 * j * pi)/n)) ** x;
plot((f(x)).real(), (x, 0, n/2)).show()
```
